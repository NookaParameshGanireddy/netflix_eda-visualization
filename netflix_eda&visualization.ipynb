```python
fig = px.pie(df , 'type' , title = 'The biggest show type')
fig.show()
```

```python
px.histogram(data_frame=df,x='country',color='country',title='The amount of productions for countries')
```

```python
px.histogram(data_frame=df,x='rating',color='rating',title='Clarification of ratings')
```

From the above figure we noticed that 3205 show is for Mature Audiences and 2157 for may not be suitable for ages under 14.

```python
px.histogram(data_frame=df,x='rating',color='type',title='Show type and it`s rating')
```

```python
px.histogram(data_frame=df,x='rating',color='country',title="Ratings of the country's productions")
```

```python
top10_rating = df['listed_in'].value_counts().head(10)
fig = px.bar(top10_rating , x=top10_rating.index , y=top10_rating.values,title='Top 10 categories of shows')
fig.show()
```

```python
top10_director = df['director'].value_counts().drop("Not Given").head(10)
fig = px.bar(top10_director , x=top10_director.index , y=top10_director.values,title='Most productive directors',width=1000,height=400,color_discrete_sequence=["green"])
fig.show()

```

```python
top10_years = df['release_year'].value_counts().head(10)
fig = px.bar(top10_years , x=top10_years.index , y=top10_years.values,title='Top 10 years in which productions took place')
fig.show()
```

```python
top10_duration = df['duration'].value_counts().head(10)
fig = px.bar(top10_duration , x=top10_duration.index , y=top10_duration.values,title='Most 10 durations')
fig.show()
```
